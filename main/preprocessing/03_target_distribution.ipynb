```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from main.utils import data_manage_utils

# Variables
ROOT_PATH = os.path.abspath("../../")
INPUT_FOLDER = os.path.join(ROOT_PATH, "data/preprocessing/processed_files")
```

```python
X_train, y_train, X_test, y_test = data_manage_utils.load_processed_data_by_folder(os.path.join(INPUT_FOLDER, "CREG"))
y_train
```

```python
y_train_filtered = y_train[((y_train >= -60) & (y_train <= 60))]
y_train_filtered
```

```python
fig, ax = plt.subplots()

plt.hist(y_train_filtered, bins=len(y_train_filtered.unique()))

plt.title("Histogram of Target Variable ([-60;60])")
ax.set_ylabel("Count")
ax.set_xlabel("Delay (Mins)")

ax.set_xticks(list(range(-60,70,10)))

plt.show()
```
