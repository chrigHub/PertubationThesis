```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import balanced_accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
y_pred_rf = pd.read_pickle("./training_results/RF/2022_10_19-1638/y_pred_RF.pkl").to_numpy()
y_pred_bl_even = bl_even_test_df = pd.read_pickle("./training_results/BL/y_pred_BL_even.pkl").to_numpy()
y_pred_bl_dist = pd.read_pickle("./training_results/BL/y_pred_BL_dist.pkl").to_numpy()
y_true = pd.read_pickle("./processed_files/01b_test_y.pkl").to_numpy()
```

```python
y_true
```

```python
print("Baseline (Even):")
print("Balanced ACC: " +str(balanced_accuracy_score(y_true, y_pred_bl_even)))
print("Precision (Weighted): " +str(precision_score(y_true, y_pred_bl_even,average="weighted")))
print("Recall (Weighted): " +str(recall_score(y_true, y_pred_bl_even,average="weighted")))
print("F1-Score (Weighted): " +str(f1_score(y_true, y_pred_bl_even,average="weighted")))
print("\n")
print("Baseline (Distribution):")
print("Balanced ACC: " +str(balanced_accuracy_score(y_true, y_pred_bl_dist)))
print("Precision (Weighted): " +str(precision_score(y_true, y_pred_bl_dist,average="weighted")))
print("Recall (Weighted): " +str(recall_score(y_true, y_pred_bl_dist,average="weighted")))
print("F1-Score (Weighted): " +str(f1_score(y_true, y_pred_bl_dist,average="weighted")))
print("\n")
print("Random Forest:")
print("Balanced ACC: " +str(balanced_accuracy_score(y_true, y_pred_rf)))
print("Precision (Weighted): " +str(precision_score(y_true,y_pred_rf,average="weighted")))
print("Recall (Weighted): " +str(recall_score(y_true, y_pred_rf,average="weighted")))
print("F1-Score (Weighted): " +str(f1_score(y_true, y_pred_rf,average="weighted")))
print("\n")
```

```python
cm = confusion_matrix(y_true,y_pred_rf, normalize="true")
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["Early","Normal","Late"]).plot()
```

```python
cm = confusion_matrix(y_true,y_pred_bl_even, normalize="true")
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["Early","Normal","Late"]).plot()
```

```python
cm = confusion_matrix(y_true,y_pred_bl_dist, normalize="true")
ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=["Early","Normal","Late"]).plot()
```
